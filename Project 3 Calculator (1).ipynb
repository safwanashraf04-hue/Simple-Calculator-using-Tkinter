```python
from tkinter import *

window = Tk()
window.title("Calculator")

e = Entry(window, width=35, borderwidth=5)
e.grid(row=0, column=0, columnspan=4, padx=10, pady=10)

# Functions
def button_click(number):
    current = e.get()
    e.delete(0, END)
    e.insert(0, str(current) + str(number))

def button_clear():
    e.delete(0, END)

def button_operation(op):
    first_number = e.get()
    global f_num
    global operation
    operation = op
    f_num = float(first_number)
    e.delete(0, END)

def button_equal():
    second_number = e.get()
    e.delete(0, END)
    try:
        if operation == "+":
            e.insert(0, f_num + float(second_number))
        elif operation == "-":
            e.insert(0, f_num - float(second_number))
        elif operation == "*":
            e.insert(0, f_num * float(second_number))
        elif operation == "/":
            e.insert(0, f_num / float(second_number))
    except ZeroDivisionError:
        e.insert(0, "Error")

# Buttons
button_1 = Button(window, text="1", width=10, height=3, command=lambda: button_click(1))
button_2 = Button(window, text="2", width=10, height=3, command=lambda: button_click(2))
button_3 = Button(window, text="3", width=10, height=3, command=lambda: button_click(3))
button_4 = Button(window, text="4", width=10, height=3, command=lambda: button_click(4))
button_5 = Button(window, text="5", width=10, height=3, command=lambda: button_click(5))
button_6 = Button(window, text="6", width=10, height=3, command=lambda: button_click(6))
button_7 = Button(window, text="7", width=10, height=3, command=lambda: button_click(7))
button_8 = Button(window, text="8", width=10, height=3, command=lambda: button_click(8))
button_9 = Button(window, text="9", width=10, height=3, command=lambda: button_click(9))
button_0 = Button(window, text="0", width=10, height=3, command=lambda: button_click(0))
button_dot = Button(window, text=".", width=10, height=3, command=lambda: button_click("."))

button_add = Button(window, text="+", width=10, height=3, command=lambda: button_operation("+"))
button_subtract = Button(window, text="-", width=10, height=3, command=lambda: button_operation("-"))
button_multiply = Button(window, text="*", width=10, height=3, command=lambda: button_operation("*"))
button_divide = Button(window, text="/", width=10, height=3, command=lambda: button_operation("/"))
button_equal = Button(window, text="=", width=10, height=3, command=button_equal)
button_clear = Button(window, text="Clear", width=32, height=3, command=button_clear)  # Wider clear button

# Layout
button_7.grid(row=1, column=0)
button_8.grid(row=1, column=1)
button_9.grid(row=1, column=2)
button_divide.grid(row=1, column=3)

button_4.grid(row=2, column=0)
button_5.grid(row=2, column=1)
button_6.grid(row=2, column=2)
button_multiply.grid(row=2, column=3)

button_1.grid(row=3, column=0)
button_2.grid(row=3, column=1)
button_3.grid(row=3, column=2)
button_subtract.grid(row=3, column=3)

button_0.grid(row=4, column=0)
button_dot.grid(row=4, column=1)
button_add.grid(row=4, column=2)
button_equal.grid(row=4, column=3)

button_clear.grid(row=5, column=0, columnspan=4)

window.mainloop()
```
